# Sales Analysis

> Using Python, Pandas, Numpy & Matplotlib to analyze the sales data and answer business questions.

**Data Dictionary**

1. Row ID - Unique ID for each row.
2. Order ID - Unique Order ID for each Customer.
3. Order Date - Order Date of the product.
4. Ship Date - Shipping Date of the Product.
5. Ship Mode - Shipping Mode specified by the Customer.
6. Customer ID - Unique ID to identify each Customer.
7. Customer Name - Name of the Customer.
8. Segment - The segment where the Customer belongs.
9. Country - Country of residence of the Customer.
10. City - City of residence of of the Customer.
11. State - State of residence of the Customer.
12. Postal Code - Postal Code of every Customer.
13. Region - Region where the Customer belong.
14. Product ID - Unique ID of the Product.
15. Category - Category of the product ordered.
16. Sub-Category - Sub-Category of the product ordered.
17. Product Name - Name of the Product
18. Sales - Sales of the Product.
19. Quantity - Quantity of the Product.
20. Discount - Discount provided.
21. Profit - Profit/Loss incurred.

## Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Get the data

```python
sales = pd.read_csv("Dataset/Superstore.csv",encoding='latin1')
sales.head()
```

## Removing Unnecessary Columns

```python
sales.drop('Row ID',axis=1,inplace=True)
```

```python
sales.head()
```

```python
sales.shape
```

```python
for columns in sales.columns:
    print(columns)
```

## Get information about data

```python
sales.info()
```

```python
sales.describe()
```

### Check for Null values

```python
sales.isna().sum()
```

### Convert the data type

```python
sales['Order Date']=pd.to_datetime(sales['Order Date'])
sales['Ship Date']=pd.to_datetime(sales['Ship Date'])
```

```python
sales.info()
```

### Create Month-Year and Year column from Order Date

```python
sales['Year'] = sales['Order Date'].dt.year
```

```python
sales['Month-Year'] = sales['Order Date'].apply(lambda x: x.strftime('%Y-%m'))
sales.head()
```

## Group sales by months

```python
sales_by_month = sales.groupby('Month-Year').sum()['Sales'].reset_index()
```

## Visualization and Analysis

### Visualization of sales treands by months

```python
fig,ax = plt.subplots(figsize=(16,8))
ax.plot(sales_by_month['Month-Year'],sales_by_month['Sales'],color='salmon')
ax.set(title='Sales Trends by Months',
       xlabel='Sales Date',
       ylabel='Sales')
plt.xticks(rotation='90');
```

```python
fig,ax = plt.subplots(figsize=(16,8))
ax.bar(sales_by_month['Month-Year'],sales_by_month['Sales'],color='lightblue')
ax.set(title='Sales Trends by Months',
       xlabel='Sales Date',
       ylabel='Sales')
plt.xticks(rotation='90');
```

### Total sales chart by the year

```python
sns.set_style('whitegrid')
sns.barplot(x='Year',y='Sales',data=sales.reset_index(), palette='rainbow')
plt.xticks(rotation = 'vertical')
plt.title('Total sales chart by the year');
```

### Most profitable products

```python
profitable_products = pd.DataFrame(sales.groupby(['Product Name']).sum()['Profit'])
profitable_products = profitable_products.sort_values(by=['Profit'], ascending = False)
profitable_products[:10]
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(14,8))
sns.barplot(x='Product Name',y='Profit',data=profitable_products[:10].reset_index(),palette="rainbow")
plt.xticks(rotation = 'vertical')
plt.title('Most Profitable Products');
```

### Most Selling Products by Sales
